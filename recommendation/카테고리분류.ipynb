```python
import pandas as pd
from datetime import datetime
data1 = pd.read_excel('../data/음식점_세부정보_24-08-31.xlsx')
data2 = pd.read_excel('../data/숙박시설_세부정보_24-08-31.xlsx')
data3 = pd.read_excel('../data/카페_세부정보_24-08-31.xlsx')
combined_df = pd.concat([data1, data2, data3], ignore_index = True)
combined_df.head()
```

```python
df = pd.DataFrame(combined_df)

column_to_process = 'category'

result_df = pd.DataFrame(index=df.index)

for i, row in df.iterrows():
    phrases = row[column_to_process].split('>')  # >를 기준으로 1차 분리
    sub_phrases = []
    for phrase in phrases:
        sub_phrases.extend(phrase.split(','))  # ,를 기준으로 2차 분리
    sub_phrases = [sub_phrase.strip() for sub_phrase in sub_phrases]  # 앞뒤 공백 제거
    for sub_phrase in sub_phrases:
        if sub_phrase not in result_df.columns:
            result_df[sub_phrase] = 0  # 칼럼이 없으면 0으로 초기화
        result_df.at[i, sub_phrase] = 1  # 해당 칼럼에 1을 넣음

final_df = df.join(result_df)

print(final_df)
```

```python
final_df = final_df.drop(columns=['검색결과', 'description', 'telephone', 'link', 'address', 'roadAddress', 'latitude', 'longitude', '음식점', '종합생활용품', '유통', '쇼핑', '생활', '편의', '꽃집', '꽃배달'])
print(final_df)
```

```python
today_date = datetime.now().strftime('%Y-%m-%d')
filename = f'카테고리_분류_{today_date}.xlsx'
final_df.to_excel(filename, index=False)
```
