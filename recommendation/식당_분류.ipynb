```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_excel('음식점_세부정보_24-08-28.xlsx')
data.head()
```

```python
df = pd.DataFrame(data)

column_to_process = 'category'

result_df = pd.DataFrame(index=df.index)

for i, row in df.iterrows():
    phrases = row[column_to_process].split('>')  # >를 기준으로 1차 분리
    sub_phrases = []
    for phrase in phrases:
        sub_phrases.extend(phrase.split(','))  # ,를 기준으로 2차 분리
    sub_phrases = [sub_phrase.strip() for sub_phrase in sub_phrases]  # 앞뒤 공백 제거
    for sub_phrase in sub_phrases:
        if sub_phrase not in result_df.columns:
            result_df[sub_phrase] = 0  # 칼럼이 없으면 0으로 초기화
        result_df.at[i, sub_phrase] = 1  # 해당 칼럼에 1을 넣음

final_df = df.join(result_df)

print(final_df)
```

```python
final_df = final_df.drop(columns=['검색결과', 'description', 'telephone', 'link', 'address', 'roadAddress', 'latitude', 'longitude', '음식점', '종합생활용품', '유통', '쇼핑', '생활', '편의', '꽃집', '꽃배달', '슈퍼', '마트'])
print(final_df)
```

```python
final_df = final_df[~df['title'].isin(['네모네 롯데백화점 영플라자 명동점', '금성슈퍼', '화원김충희프라자'])]
```
