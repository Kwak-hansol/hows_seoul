```python
import requests
import pandas as pd
from datetime import datetime

# 클라이언트 ID와 시크릿
client_id = '2gJMWkJ1gQO98RmCtnh4'
client_secret = 'kZcfFBNrvO'

current_time = datetime.now().strftime('%y-%m-%d')

def search_location(query):
    """주어진 쿼리로 네이버 API를 사용하여 위치 정보를 검색합니다."""
    url = "https://openapi.naver.com/v1/search/local.json"
    headers = {
        "X-Naver-Client-Id": client_id,
        "X-Naver-Client-Secret": client_secret
    }
    params = {
        "query": query,
        "display": 1,  # 첫 번째 결과만 가져오기
        "sort": "random"  # 정렬 방식 (필요시 변경)
    }
    
    response = requests.get(url, headers=headers, params=params)
    
    if response.status_code == 200:
        data = response.json()
        items = data.get('items', [])
        if items:
            # 첫 번째 항목의 위치 정보를 반환
            location = items[0]
            return location
        else:
            return None
    else:
        print(f"Error: {response.status_code}")
        response.raise_for_status()

def search_locations(queries, original_df):
    """여러 장소를 검색하고 각 장소의 정보를 테이블로 반환합니다."""
    results = []
    for idx, query in enumerate(queries):
        location = search_location(query)
        
        if location:
            # 위도와 경도를 문자열로 변환한 후 형식 변환
            latitude = location['mapy']
            longitude = location['mapx']
            latitude = float(latitude[:2] + '.' + latitude[2:])
            longitude = float(longitude[:3] + '.' + longitude[3:])
            
            # 데이터를 딕셔너리로 정리
            location_data = {
                '검색어': original_df.loc[idx, '검색어'],  # 원본 데이터에서 검색어 가져오기
                '검색결과': query,
                'title': location.get('title', '').replace('<b>', '').replace('</b>', ''),
                'link': location.get('link', ''),
                'category': location.get('category', ''),
                'description': location.get('description', ''),
                'telephone': location.get('telephone', ''),
                'address': location.get('address', ''),
                'roadAddress': location.get('roadAddress', ''),
                'latitude': latitude,
                'longitude': longitude
            }
            
            # 결과 리스트에 추가
            results.append(location_data)
            print(f"'{query}' 정보 추가 완료.")
        else:
            print(f"'{query}' 위치를 찾을 수 없습니다.")
    
    # 결과를 DataFrame으로 변환
    df = pd.DataFrame(results)
    return df

# 엑셀 파일에서 '검색어'와 '검색결과' 컬럼의 데이터 읽어오기
# input_file = "../data/음식점_검색결과_24-08-31.xlsx"
# input_file = "../data/숙박시설_검색결과_24-08-31.xlsx"
input_file = "../data/카페_검색결과_24-08-31.xlsx"
df_queries = pd.read_excel(input_file)

# '검색결과' 컬럼을 리스트로 변환
queries = df_queries['검색결과'].dropna().tolist()

# 장소 검색 및 테이블 생성
locations_df = search_locations(queries, df_queries)

# 테이블을 Excel 파일로 저장
# output_file = f"../data/음식점_세부정보_{current_time}.xlsx"
# output_file = f"../data/숙박시설_세부정보_{current_time}.xlsx"
output_file = f"../data/카페_세부정보_{current_time}.xlsx"
locations_df.to_excel(output_file, index=False)
print(f"장소 정보가 '{output_file}' 파일에 저장되었습니다.")
```
