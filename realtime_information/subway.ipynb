```python
import requests
import xml.etree.ElementTree as ET
import time

def get_request_url():
    base_url1 = 'http://swopenAPI.seoul.go.kr/api/subway'
    access_key = '58427970596c6f65353957704f5644'
    type = 'xml'
    base_url2 = 'realtimeStationArrival'
    start_row = 0
    end_row = 5
    location = '서울'  # 예시: 서울역 (역 빼고)

    # 완성된 URL
    url = f'{base_url1}/{access_key}/{type}/{base_url2}/{start_row}/{end_row}/{location}'
    return url


def fetch_subway_data():
    url = get_request_url()
    response = requests.get(url)
    if response.status_code == 200:
        return response.content  # XML 데이터를 반환
    else:
        print("API 요청 실패:", response.status_code)
        return None


def parse_and_display_data(xml_data):
    root = ET.fromstring(xml_data)

    print(f"실시간 지하철 도착 정보 ({time.strftime('%Y-%m-%d %H:%M:%S')}):")
    print("-" * 50)

    for row in root.findall('.//row'):
        subway_id = row.find('subwayId').text
        train_line_nm = row.find('trainLineNm').text
        station_name = row.find('statnNm').text
        arrival_message = row.find('arvlMsg2').text
        train_status = row.find('btrainSttus').text
        reception_time = row.find('recptnDt').text

        print(f"지하철 ID: {subway_id}")
        print(f"열차 노선: {train_line_nm}")
        print(f"역 이름: {station_name}")
        print(f"도착 정보: {arrival_message}")
        print(f"열차 상태: {train_status}")
        print(f"수신 시간: {reception_time}")
        print("-" * 50)


def main():
    while True:
        xml_data = fetch_subway_data()
        if xml_data:
            parse_and_display_data(xml_data)
        time.sleep(10)  # 10초마다 갱신


if __name__ == "__main__":
    main()
```
